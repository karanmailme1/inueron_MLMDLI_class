Note: BS is based on Divide and Conquer (DAC) approach 

```python
def BinarySearch(ip, low_i, high_i, x):
    
    if low_i <= high_i:      ##Small Problem -> O(1) # <= handles boundary level case when x is less than lowest element in array
        if ip[low_i] == x:   # this case runs only when there is one element in array
            print(x, ' is in array at index::', low_i)
            return low_i
        
        else:
            print(x, ' is not part of array:', ip)
            return -1       ## Element not present in array at end of recursion
    
    else:
        mid_i = low_i + (high_i - low_i ) // 2     ## Divide -> O(1)
        ##Note: we can also do 'mid = (high + low) // 2' above but in some boundry level cases, 
        ##      it may cause overflow problem for a very large array with boundry value high and low may hold
        
        if (ip[mid_i] == x):  #runs when mid element matches x 
            print(x, ' is in array at index:', mid_i)
            return mid_i
        
        elif (ip[mid_i] > x): # search on left side of array
            #print(mid_i, 'left')
            return BinarySearch(ip, low_i, (mid_i-1), x)       ##Conquer: Recursive call -> T(n/2)
        
        elif (ip[mid_i] < x): # search on right side of array
            #print(mid_i,'right')
            return BinarySearch(ip, (mid_i+1), high_i, x)      ##Conquer: Recursive call -> T(n/2)
```

```python
input_arr = [3,6,8,11,23,35,45,67,70,89,90]
#input_arr = [1]
print(len(input_arr)//2)
#print(input_arr[0:3])
```

```python
x = 100
lower_index = 0
higher_index = (len(input_arr)-1)

BinarySearch(input_arr, lower_index, higher_index, x)
```
