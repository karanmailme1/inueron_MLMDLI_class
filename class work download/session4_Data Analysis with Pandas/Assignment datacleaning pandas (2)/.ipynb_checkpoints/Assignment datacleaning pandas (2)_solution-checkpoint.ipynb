```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})

print(df.info())
df
```

```python
# 1. Some values in the the FlightNumber column are missing. 
# These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place. 
# Fill in these missing numbers and make the column an integer column (instead of a float column).

pd.set_option('mode.chained_assignment', None) #to supress warning

df['FlightNumber'][df['FlightNumber'].isnull()] = [10055,10075]
df['FlightNumber'] = df['FlightNumber'].astype('int')

pd.set_option('mode.chained_assignment', 'raise')

df
```

```python
# 2. The From_To column would be better as two separate columns! 
# Split each string on the underscore delimiter _ to give a new temporary DataFrame with the correct values. 
# Assign the correct column names to this temporary DataFrame.

df_temp = df['From_To'].str.split('_', expand=True)
df_temp.columns = ['From','To']
df_temp
```

```python
# 3. Notice how the capitalisation of the city names is all mixed up in this temporary DataFrame. 
# Standardise the strings so that only the first letter is uppercase (e.g. "londON" should become "London".)

#'AbC'.capitalize()
#df_temp['From'].str.capitalize()

df_temp = df_temp.apply(lambda x: x.str.capitalize())
df_temp
```

```python
# 4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.

df.drop(columns=['From_To'], inplace=True)
df = pd.concat([df_temp, df], axis=1)
df
```

```python
# 5. In the RecentDelays column, the values have been entered into the DataFrame as a list. 
# We would like each first value in its own column, each second value in its own column, and so on. 
# If there isn't an Nth value, the value should be NaN.
# Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. 
# and replace the unwanted RecentDelays column in df with delays.

RecentDelays = df['RecentDelays']
print(RecentDelays)

#split list column into individual lists of columns based on index
RecentDelays = list(RecentDelays.str)
print(RecentDelays)

df.drop(columns=["RecentDelays"], inplace=True)
#df.concat([df, RecentDelays], axis=1)
#df
```

```python
#convert list into dict with keys as required column headers

rd={}
for i,j in enumerate(RecentDelays):
    rd['delay_'+str(i)] = j
print(rd)

rd = pd.DataFrame(rd)
print(rd)
```

```python
#concatinate df and rd

df = pd.concat([df,rd], axis=1)
df
```
